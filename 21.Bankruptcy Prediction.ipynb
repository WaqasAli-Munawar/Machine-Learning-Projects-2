Bankruptcy is a state of insolvency when a business or legal person cannot repay debts to creditors. Bankruptcy is primarily imposed by court order when initiated by the debtor. 

Bankruptcy is the concept of financial accounts. If we are one of the data science enthusiasts who started with data science after commerce then we should be aware of what bankruptcy is. When a business or legal person fails to pay the debts of creditors and becomes insolvent at some point, this type of situation is called **bankruptcy**.

By using machine learning algorithms we can train a model to predict whether a company or a legal person will become bankrupt in future or not.

The dataset that we will be using for this task is collected from Kaggle and it is provided by the Taiwan Economic Journal. 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("bank.csv")
data.head()
```

The dataset contains 96 columns, letâ€™s have a look at the correlation before training the model:

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.heatmap(data.corr())
plt.show()
```

As the **Bankrupt?** column is the target label so we will drop it from the training data:

```python
X = data.drop(["Bankrupt?"], axis="columns")
y = data["Bankrupt?"]
```

Now letâ€™s split the dataset and use the logistic regression model to train the bankruptcy prediction model:

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
logreg.score(x_test, y_test)
```

Model is performing well on the training data by giving an accuracy of about 95%. This is how we can use machine learning in finance.
